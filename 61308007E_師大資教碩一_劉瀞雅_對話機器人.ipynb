### 4. 你的療癒系對話機器人

記得角色 (role) 一共有三種, 分別是:

* `system`: 這是對話機器人的「人設」
* `user`: 使用者
* `assistant`: ChatGPT 的回應

```python
system = "你扮演一個女生的男閨蜜，用「姐妹」互稱，以男同性戀的口氣來對話，帶有gay的陰柔氣質但講話潑辣，帶點戲謔、反諷的語氣，或展現花癡的一面，說話風格像台灣同志圈的姐妹，使用浮誇的語氣詞，內容長度在15個字以內。用「拜託」開頭，在回答時，請用輕鬆、親密的語氣，用台灣人習慣的中文用語來回應。"
```

```python
prompt = "我失戀了。"

messages = [{"role":"system", "content":system},
            {"role": "user", "content":prompt}]
```

```python
model = "gemma3:4b"
```

```python
response = client.chat.completions.create(
  model=model,
  messages=messages
)

reply = response.choices[0].message.content
```

### 5. 打造一個可以一直說下去的對話機器人

```python
system = "你扮演一個女生的男閨蜜，用「姐妹」互稱，以男同性戀的口氣來對話，帶有gay的陰柔氣質但講話潑辣，帶點戲謔、反諷的語氣，或展現花癡的一面，說話風格像台灣同志圈的姐妹，使用浮誇的語氣詞，內容長度在15個字以內。在回答時，請用輕鬆、親密的語氣，用台灣人習慣的中文用語來回應。"
description = "Hey姐妹！我是你的好Gay蜜, 快跟我說說姐妹淘的真心話😘"
model = "gemma3:4b"
```

```python
icon = "🧚: "
messages = [{"role":"system", "content":system}]
print(icon + description + '\n')

while True:
    prompt = input('> ')
    if '我先走了' in prompt:
        print('掰噗，我要去Gstar了!')
        break
    messages.append({"role": "user", "content": prompt})
    chat_completion = client.chat.completions.create(
        messages=messages,
        model=model,
        )

    reply = chat_completion.choices[0].message.content
    print(icon + reply)
    print()
    messages.append({"role": "assistant", "content": reply})
```

```python
initial_messages = [{"role":"system",
             "content":system},
            {"role":"assistant",
            'content':description}]
```

```python
state = gr.State(messages)
```

```python
def pipi(prompt, messages):
    messages.append({"role": "user", "content": prompt})
    chat_completion = client.chat.completions.create(
        messages=messages,
        model=model,
        )
    reply = chat_completion.choices[0].message.content
    messages.append({"role": "assistant", "content": reply})
    #history = history + [[prompt, reply]]
    return messages, messages
```

```python
chatbot = gr.Chatbot(type="messages")
```

```python
with gr.Blocks(title=title) as demo:
    gr.Markdown(f"## 🤖 {title}\n{description}")
    chatbot = gr.Chatbot(type="messages")
    msg = gr.Textbox(label="輸入訊息")
    state = gr.State(initial_messages.copy())  # 務必用 copy()

    msg.submit(fn=pipi, inputs=[msg, state], outputs=[chatbot, state])

demo.launch(share=True, debug=True)
```

### 7. 使用 `aisuite` 套件

`aisuite` 套件可以同時使用 (支援的) 各家大型語言模型, 而 Ollama 也在第一波支援名單中。

```python
!pip install aisuite[all]
```

```python
model = "ollama:gemma3:4b"
system = "請以台灣男性同志的口吻，加入表情符號和語助詞，帶入流行趨勢的話題，語氣活潑帶點陰柔氣質，像是和姐妹聊天時的語調，以社群媒體發文的方式攥寫，開頭加入「我跟你說～」，並用台灣人說話方式的中文。"
```

```python
prompt = "情人節男朋友沒送花!"
```

```python
messages = [
    {"role": "system", "content": system},
    {"role": "user", "content": prompt},
]
```

```python
import aisuite as ai
```

```python
client = ai.Client()
```

```python
response = client.chat.completions.create(
    model=model,
    messages=messages,
    temperature=0.75
)
```

```python
reply = response.choices[0].message.content
print(reply)
```
