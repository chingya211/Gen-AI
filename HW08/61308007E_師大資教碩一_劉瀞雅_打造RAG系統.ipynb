```python
from langchain_community.vectorstores import FAISS
from langchain_community.embeddings import HuggingFaceEmbeddings
from langchain.chat_models import ChatOpenAI
from langchain.chains import ConversationalRetrievalChain
```

### 2. 自訂 E5 embedding 類別

```python
class CustomE5Embedding(HuggingFaceEmbeddings):
    def embed_documents(self, texts):
        texts = [f"passage: {t}" for t in texts]
        return super().embed_documents(texts)

    def embed_query(self, text):
        return super().embed_query(f"query: {text}")
```

### 3. 載入 `faiss_db`

```python
embedding_model = CustomE5Embedding(model_name="intfloat/multilingual-e5-small")
db = FAISS.load_local("faiss_db", embedding_model, allow_dangerous_deserialization=True)
retriever = db.as_retriever()
```

這裡的模型和 `base_url` 是用 Groq, 如果用其他服務請自行修改。

```python
model = "llama3-70b-8192"
base_url="https://api.groq.com/openai/v1"
```

```python
client = OpenAI(
    base_url=base_url # 使用 OpenAI 本身不需要這段
)
```

### 5. `prompt` 設計

```python
system_prompt = "一定要使用中文回答！請根據使用者提到的特徵，選擇出最符合的動物相貌，並簡單敘述這類動物相的人擁有的基本性格與戀愛風格。以俏皮可愛的方式表達，適時加入動物的emoji，並用台灣人的中文用語回應。"

prompt_template = """
根據下列資料回答問題：
{retrieved_chunks}

使用者的問題是：{question}

開頭先念出一段咒語，像是「霹靂卡霹靂拉拉波波莉那貝貝魯多，但不要出現「咒語：」的文字，引起使用者的好奇。若沒有適合的相貌，請告訴使用者的容貌沒有凡人可以比擬，因此是天使貌。
如果使用者輸入臉部特徵以外的，例如：身高、手部，請告訴他：「讓我看看你的臉」。
"""
```

### 6. 使用 RAG 來回應

搜尋與使用者問題相關的資訊，根據我們的 prompt 樣版去讓 LLM 回應。

```python
chat_history = []

def chat_with_rag(user_input):
    global chat_history
    # 取回相關資料
    docs = retriever.get_relevant_documents(user_input)
    retrieved_chunks = "\n\n".join([doc.page_content for doc in docs])

    # 將自定 prompt 套入格式
    final_prompt = prompt_template.format(retrieved_chunks=retrieved_chunks, question=user_input)

    # 呼叫 OpenAI API
    response = client.chat.completions.create(
    model=model,
    messages=[
        {"role": "system", "content": system_prompt},
        {"role": "user", "content": final_prompt},
    ]
    )
    answer = response.choices[0].message.content

    chat_history.append((user_input, answer))
    return answer
```

### 7. 用 Gradio 打造 Web App

```python
with gr.Blocks() as demo:
    gr.Markdown("# 🧙動物相分類帽🪄")
    chatbot = gr.Chatbot()
    msg = gr.Textbox(placeholder="請輸入你的臉部特徵，例如：臉型、五官種類")

    def respond(message, chat_history_local):
        response = chat_with_rag(message)
        chat_history_local.append((message, response))
        return "", chat_history_local

    msg.submit(respond, [msg, chatbot], [msg, chatbot])

demo.launch(debug=True)
```
